```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv("vpn.log")
df
```

```python
df = df.drop(columns=["Node", "Active", "Service", "Error", "Port", "Proto"])
df[["Duration"]] = df[["Duration"]].fillna(value=0)
```

```python
df_sum = df.groupby(["Username"]).sum()
df_mean = df.groupby(["Username"]).mean()
```

```python
df_sum["average bytes"] = df_sum['Bytes Total'] / df_sum["Duration"]
df_sum["bytes / auth"] = df_sum["Bytes Total"] / df_sum["Auth"]
df_mean["average bytes"] = df_mean['Bytes Total'] / df_mean["Duration"]
df_mean["bytes / auth"] = df_mean["Bytes Total"] / df_mean["Auth"]
```

```python
df_mean.sort_values("bytes / auth", ascending=False)
```

```python
df_sum.sort_values("Duration", ascending=False)
```
